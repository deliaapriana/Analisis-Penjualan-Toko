```python
# Langkah 1: Import library
import pandas as pd
import matplotlib.pyplot as plt

# Langkah 2: Buat data penjualan sederhana (dummy)
data = {
    'Tanggal': ['2025-01-10', '2025-01-12', '2025-01-15', '2025-02-01', '2025-02-05', '2025-02-10', '2025-03-02'],
    'Produk': ['Lip Tint', 'Serum', 'Lip Tint', 'Masker', 'Serum', 'Lip Tint', 'Masker'],
    'Harga': [45000, 120000, 45000, 25000, 120000, 45000, 25000],
    'Jumlah': [3, 2, 4, 10, 1, 5, 8]
}

# Langkah 3: Ubah ke DataFrame (struktur tabel)
df = pd.DataFrame(data)

# Langkah 4: Hitung total pendapatan
df['Pendapatan'] = df['Harga'] * df['Jumlah']

# Tampilkan tabel
df

```

```python
# Total pendapatan per produk
pendapatan_per_produk = df.groupby('Produk')['Pendapatan'].sum()

# Buat grafik batang
pendapatan_per_produk.plot(kind='bar')

plt.title('Total Pendapatan per Produk')
plt.xlabel('Produk')
plt.ylabel('Total Pendapatan (Rp)')
plt.grid(True)
plt.show()
```

```python
# Simpan data ke file CSV
df.to_csv('penjualan_toko_dummy.csv', index=False)

print("File CSV berhasil dibuat!")
```
